```python
data=data[['Humidity','Temperature (C)']]
data.columns=['Humidity','Temperature']
data
```

```python
train_data,test_data= train_test_split(data,test_size=3,shuffle=True,random_state=42)


train_x, train_y = np.array(train_data['Humidity']).reshape(-1,1), np.array(train_data['Temperature']).reshape(-1,1)
test_x, test_y = np.array(test_data['Humidity']).reshape(-1,1), np.array(test_data['Temperature']).reshape(-1,1)


test_x.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
train_x=scaler.fit_transform(train_x)

model=LinearRegression()
model.fit (train_x,train_y)
```

```python
model.score(train_x,train_y)
```

```python
score=cross_val_score(model,train_x,train_y,scoring='neg_mean_absolute_error',cv=3)

score=(-score)

score.mean()
```

```python
test_x.shape
```

```python
test_x=scaler.transform(test_x)
test_predict=model.predict(test_x)
score=mean_absolute_error(test_y,test_predict)
score
```
